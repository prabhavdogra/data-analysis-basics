```python
from matplotlib import pyplot as plt
plt.plot([1,2,3],[2,5,7])
plt.show()
```

# How to add title and lables to x and y axis

```python
plt.plot([1,2,3],[2,5,7])
plt.title("Raghav ")
plt.ylabel("this is Y ------>")
plt.xlabel("this is X ------>")
plt.show()
```

# How to import two linw datas in one plot

```python
x = [1,5,4]
y = [23 , 45 , 89]
x2 = [2,6,5]
y2 = [4,2,9]
plt.plot(x , y)
plt.plot(x2 , y2)
plt.show()
```

# how to change plot style

```python
from matplotlib import style
style.use('ggplot')
x = [1,5,4]
y = [23 , 45 , 89]
x2 = [2,6,5]
y2 = [4,2,9]
plt.plot(x , y)
plt.plot(x2 , y2)
plt.show()
```

```python
x = [1,5,4]
y = [23 , 45 , 89]
x2 = [2,6,5]
y2 = [4,2,9]
plt.plot(x , y , label='first')
plt.plot(x2 , y2 , label = 'second')
plt.legend()
plt.show()
```

```python
x = [1,5,4]
y = [23 , 45 , 89]
x2 = [2,6,5]
y2 = [4,2,9]
plt.plot(x , y , label='first' , linewidth = 18)
plt.plot(x2 , y2 , label = 'second')
plt.legend()
plt.show()
```
